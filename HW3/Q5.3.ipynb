```python
import numpy as np
import matplotlib.pyplot as plt
```

1) Plot leverage h_{ii} vs. absolute value of (x_{i} - x-mean) for p=1. 

```python
np.random.seed(42)
n = 100
p_1 = 1

X_1= np.random.randn(n, p_1)
X_1_mean = np.mean(X_1, axis=0)
X_SLR = np.insert(X_1, 0, 1, axis=1)

H_SLR = np.dot(X_SLR, np.linalg.inv(np.dot(X_SLR.T, X_SLR))).dot(X_SLR.T)
h_ii_SLR = np.diag(H_SLR)

absolute_difference = np.abs(X_1[:, 0] - X_1_mean[0]) 

plt.scatter(absolute_difference, h_ii_SLR, color='b')
plt.xlabel('|x_i - x_mean|')
plt.ylabel('h_ii')
plt.title('h_ii vs. |x_i - x_mean| for p = 1')
```

2) Plot leverage h_{ii} vs. 2-norm of (x_i - x_mean) for p=4.

```python
p_2 = 4

X_2 = np.random.randn(n, p_2)
X_2_mean = np.mean(X_2, axis=0)
X_MLR = np.insert(X_2, 0, 1, axis=1)

H_MLR = np.dot(X_MLR, np.linalg.inv(np.dot(X_MLR.T, X_MLR))).dot(X_MLR.T)
h_ii_MLR = np.diag(H_MLR)

norm_difference = np.linalg.norm(X_2 - X_2_mean, axis=1)

plt.scatter(norm_difference, h_ii_MLR, color='r')
plt.xlabel('||x_i - x_mean||')
plt.ylabel('h_ii')
plt.title('h_ii vs. ||x_i - x_mean|| for p > 1')
```
